```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# training_data features
data_features = pd.read_csv("train_logs.csv")
#training_data scores
data_scores = pd.read_csv("train_scores.csv")
data = pd.merge(data_features, data_scores, on='id', how='left')
```

```python
def plot_feature_distribution(df, column_name, feature_type):
    """
    Plots the distribution of a dataframe feature.
    
    :param df: pandas DataFrame containing the data
    :param column_name: the name of the column to plot
    :param feature_type: the type of the feature ('numerical' or 'categorical')
    """
    # Plot based on the specified feature type
    if feature_type == 'numerical':
        # Use a histogram or boxplot for numerical features
        sns.histplot(df[column_name], kde=False)
        plt.title(f'Distribution of {column_name} (Numerical)')
        plt.xlabel(column_name)
        plt.ylabel('Frequency')
    elif feature_type == 'categorical':
        # Use a count plot for categorical features
        sns.countplot(y=column_name, data=df, order=df[column_name].value_counts().index)
        plt.title(f'Distribution of {column_name} (Categorical)')
        plt.xlabel('Count')
        plt.ylabel(column_name)
    else:
        raise ValueError("feature_type must be 'numerical' or 'categorical'")

    plt.tight_layout()
    plt.show()
```

```python
plot_feature_distribution(data, 'cursor_position', 'numerical')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

def plot_distributions(df):
    for column in df.columns:
        plt.figure(figsize=(10,4))
        
        # Check data type of the column
        if df[column].dtype == 'object':  # Categorical data
            sns.countplot(y=column, data=df, order=pd.value_counts(df[column]).iloc[:10].index)
            plt.title(f'Distribution of {column} - Categorical')
        else:  # Numerical data
            sns.histplot(df[column], kde=False)
            plt.title(f'Distribution of {column} - Numerical')
        
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()

# Call the function with your dataframe
plot_distributions(data)
```
